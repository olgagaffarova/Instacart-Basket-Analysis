# 01 Importing Libraries

```python
import pandas as pd
import numpy as np
import os 
```

# 02 Importing Data

```python
path = r'/Users/woodoooo/Desktop/Instacart Basket Analysis'

df_orders_products_combined = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data','orders_products_combined.pkl'))

df_products = pd.read_csv(os.path.join(path,'02 Data','Prepared Data','products_cleaned.csv'), index_col = False)
```

```python
# Drop unnecessary columns
df_orders_products_combined = df_orders_products_combined.drop(['Unnamed: 0'], axis=1)
```

```python
# Drop unnecessary columns
df_products = df_products.drop(['Unnamed: 0'], axis=1)
```

# 03 Merging dataframes

```python
# Sample merge test of two DataFrames — df_orders_products_combined and df_products 
# we should use the product_id column, which is common to both.

sample = df_orders_products_combined.sample(n=100_000)
pd.merge(sample, df_products, on='product_id', how='inner', indicator='match')
```

```python
# actual merge of 'orders_products_combined' and '_df_pords'
ords_prods_merge = df_orders_products_combined.merge(df_products, on = ['product_id'], how='inner', indicator = 'match' )
```

```python
ords_prods_merge.shape
```

```python
# drop the flag column '_merge'

ords_prods_merge.drop('_merge', axis=1, inplace=True)
```

# 04 Exporting Data

```python
# Export data to pkl
ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merge.pkl'))
```
